```
import pandas as pd
import numpy as np
import matplotlib as mpl
```

```
f = pd.read_csv("https://gist.githubusercontent.com/linhnd99/9b5c10be3fb989e2ccbe393927a10eeb/raw/9f472c009a9b584c031d3fde57e03e57e292d109/sale.csv")
X = f.values[:,1:4]
Y = f.values[:,4]
n=len(X)
f
```

```
def f(w1,w2,w3,tv,radio,news):
  return w1*tv+w2*radio+w3*news
```

```
def MSE(w1,w2,w3):
  mse=0
  for tv,rad,news,y_i in zip(X[:,0],X[:,1],X[:,2],Y):
    mse += (y_i - f(w1,w2,w3,tv,rad,news))**2
  return mse/(2*n)
```

```
def gradient_w1(w1,w2,w3):
  ans = 0.0
  for tv,rad,news,y_i in zip(X[:,0],X[:,1],X[:,2],Y):
    ans += -tv*(y_i-f(w1,w2,w3,tv,rad,news))
  return -2/(len(X))*ans

def gradient_w2(w1,w2,w3):
  ans=0.0
  for tv,rad,news,y_i in zip(X[:,0],X[:,1],X[:,2],Y):
    ans += -rad*(y_i-f(w1,w2,w3,tv,rad,news))
  return -2/(len(X))*ans

def gradient_w3(w1,w2,w3):
  ans=0.0
  for tv,rad,news,y_i in zip(X[:,0],X[:,1],X[:,2],Y):
    ans += -news*(y_i-f(w1,w2,w3,tv,rad,news))
  return -2/(len(X))*ans
```

```
learning_rate = 0.0001
epochs=100
w1 = 1.0
w2 = 1.0
w3 = 1.0
for i in range(epochs):
  cw1=w1
  cw2=w2
  cw3=w3
  w1 = w1 - learning_rate*gradient_w1(cw1,cw2,cw3)
  w2 = w2 - learning_rate*gradient_w2(cw1,cw2,cw3)
  w3 = w3 - learning_rate*gradient_w3(cw1,cw2,cw3)
  print("MSE {}th: {}".format(i+1,MSE(w1,w2,w3)))
  print("\tw1: {} \n\tw2: {} \n\tw3: {}\n".format(w1,w2,w3))
```
